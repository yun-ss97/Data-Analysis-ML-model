##### Ⅲ. Overfitting 문제
(1) Overfitting 의 정의

overfitting 이란, training data 에서만 가지고 있는 특징까지 지나치게 학습하여, 새로운 데이터를 제대로 분류하지 못하는 경우를 말한다. overfitting 이 일어난 경우, training data 에 대해서는 일정 수준 이상의 예측 정확도를 보이지만, valid data 나 test data 를 적용하였을 때는 성능이 떨어진다. 

(2) Overfitting 발생원인

training data 와 test data 가 중복될 때 생길 수 있다. 또한 feature 의 개수가 많거나 가지고 있는 데이터의 수가 너무 적어서 발생할 수도 있다. 

(3) Overfitting 해결방법

overfitting 을 해결하는 방법으로는 학습데이터를 추가적으로 수집하기, feature 개수 줄이기, 정규화(regularization)가 있다. 정규화는 classifier 가 지나치게 데이터에 overfit 되어 그래프가 구불구불하게 되는 것을 방지해준다. 덧붙여서, 학습데이터와 평가데이터가 중복되는 경우는 이 프로젝트에서는 발생하지 않는다. 앞서 전처리 과정에서 train, validation, test 데이터를 나누고 시작했기 때문이다.

위의 방법들 중 학습데이터를 추가적으로 수집하는 것은 어려우므로, feature 개수 줄이기와 정규화하는 방법으로 overfitting 문제를 해결해보고자 한다. feature selection 의 경우 데이터 전처리에 해당하는 과정이기도 하므로, feature를 고른 이후, 7가지 전체 모델에서의 성능 변화와 ovefit 여부를 살펴볼 것이다.

##### Ⅳ. 선정한 모델의 성능 향상
선정한 모델의 성능을 향상시키기 위해 설정에 몇 가지 변화를 주었다.
activation function은  input 을 output 으로 변환하는 과정이 각각 다르기 때문에 성능에도 변화가 있다. activation function 에 따라 달라지는 성능을 확인해보았다. 

또한, layer의 수, batch_size, epoch 의 값을 변화하였다. 이 값들을 변화시킨 이유는 성능과 overfitting 모두와 높은 관련성이 있기 때문이다. 무조건 값을 올려주면 성능이 향상되는 것이 아니며, training data 에서 성능이 향상되더라도, overfitting 의 가능성이 높아질 수 있다. overfitting 이 되지 않으면서 성능을 가장 향상시킬 수 있는 모델의 조건을 탐색해보았다.

(1) activation function 의 변화(sigmoid→tanh→ReLU)

-sigmoid: 결과값이 0과 1 사이로 제한된다. gradient vanishing 문제가 발생할 수 있다.

-tanh: 결과값이 -1과 1 사이로 제한된다. 여전히 gradient vanishing 문제가 생길 수 있다.

-ReLU(Rectified Linear Unit): input 이 음수일 경우 0을 출력하고, input 이 양수일 경우, output 을 그대로 출력한다. 최근 가장 많이 쓰이는 활성함수이다.  

(2) layer 의 수(2→3→4): hidden layer 개수를 의미한다.

(3) batch_size(5→10→50): 

한 번 학습 시, 몇 개의 데이터를 가지고 학습할 것인지를 의미한다. 즉, 한번의 기울기 갱신에 몇 개의 train data 를 사용할 것인지에 대한 값이다. 

(4) epoch(100→500): 학습의 횟수를 의미한다.

### Ⅰ. 데이터 확인 및 전처리-(4) Confusion Matrix 함수

```python
def print_score(classifier,X_train,y_train,X_valid,y_valid,train=True):
    if train == True:
        print("[Training results]:\n")
        print('Accuracy Score: {0:.4f}\n'.format(accuracy_score(y_train,classifier.predict(X_train))))
        print('Classification Report:\n{}\n'.format(classification_report(y_train,classifier.predict(X_train))))
        print('Confusion Matrix:\n{}\n'.format(confusion_matrix(y_train,classifier.predict(X_train))))
        res = cross_val_score(classifier, X_train, y_train, cv=10, n_jobs=-1, scoring='accuracy')
        print('Average Accuracy:\t{0:.4f}\n'.format(res.mean()))
        print('Standard Deviation:\t{0:.4f}'.format(res.std()))
    #elif train == False:
        print("\n")
        print("\n")
        print("[Valid results]:\n")
        print('Accuracy Score: {0:.4f}\n'.format(accuracy_score(y_valid,classifier.predict(X_valid))))
        print('Classification Report:\n{}\n'.format(classification_report(y_valid,classifier.predict(X_valid))))
        print('Confusion Matrix:\n{}\n'.format(confusion_matrix(y_valid,classifier.predict(X_valid))))
```

이후 training data 와 valid data 의 성능을 평가하기 위해 confusion matrix 를 나타내는 함수를 정의하였다. matrix 내부는 수업시간에 배웠던 accuracy, precision, recall, f1-score 로 구성하였다.
1. accuracy: 전체 데이터에서 올바르게 클래스를 분류한 개수의 비율이다.
2. precision: positive predictive value 로서, positive 라고 예측한 것 중에서 몇 개를 맞추었는지를 나타낸다.
3. recall: 실제 positive class 에서 몇 개를 positive 로 맞추었는지에 대한 비율이다.
4. f1-score: precison 과 recall 의 harmonic mean 이다. f1-score 가 높을수록 분류모델의 예측력이 좋다는 것을 의미한다.

## Ⅲ. overfitting 확인 및 문제해결

### Ⅲ. Overfitting-(1) Overfitting 의 정의

overfitting 이란, training data 를 지나치게 학습하여, 새로운 데이터를 제대로 분류하지 못하는 경우를 말한다. overfitting 이 일어난 경우, training data에 대해서는 일정 수준 이상의 예측 정확도를 보이지만, valid data 나 test data 를 적용하였을 때는 성능이 떨어진다. 

### Ⅲ. Overfitting-(2) Overfitting 발생원인

training data 와 test data 가 중복될 때 생길 수 있다. 또한 feature 의 개수가 많거나 가지고 있는 데이터의 수가 너무 적어서 발생할 수도 있다. 

### Ⅲ. Overfitting-(3) Overfitting 해결방법

overfitting 을 해결하는 방법으로는 학습데이터를 추가적으로 수집하기, feature 개수 줄이기, 정규화(regularization)가 있다. 정규화는 classifier 가 지나치게 데이터에 overfit 되어 그래프가 구불구불하게 되는 것을 방지해준다. 

위의 방법들 중 학습데이터를 추가적으로 수집하는 것은 어려우므로, feature 개수 줄이기와 정규화하는 방법으로 overfitting 문제를 해결해보고자 한다. feature selection 의 경우 데이터 전처리에 해당하는 과정이기도 하므로, feature를 고른 이후, 7가지 전체 모델에서의 성능 변화와 overfit 여부를 살펴볼 것이다. 전처리 과정에서 train, valid, test 데이터를 나누었기 때문에 데이터가 중복되어서 생기는 overfitting 은 발생하지 않는다.

#### 방법2. Pearson 의 Correlation

feature 간의 상관관계를 파악하여 feature 를 고르는 방법이다.

```python
feature_name = X.columns.tolist()
def cor_selector(X, y):
    cor_list = []
    for i in X.columns.tolist():
        cor = np.corrcoef(X[i], y)[0, 1]
        cor_list.append(cor)
    cor_list = [0 if np.isnan(i) else i for i in cor_list]
    cor_feature = X.iloc[:,np.argsort(np.abs(cor_list))[-10:]].columns.tolist()
    cor_support = [True if i in cor_feature else False for i in feature_name]
    return cor_support, cor_feature
```

고르는 feature 의 개수를 임의로 10개로 잡았다.

```python
cor_support, cor_feature = cor_selector(X, y)
print(str(len(cor_feature)), 'selected features')
```

L2 Normalizer 를 사용하였을 때 overfitting 문제가 해결되는 것을 관찰했다. Dropout 에서는 전체적으로 성능이 낮아졌는데, 이것은 Dropout의 작동원리를 고려하였을 때, 이 실험에서는 layer 수가 너무 적기 때문에 오히려 성능이 하락했다고 추측해보았다. Dropout 기법은 layer 수가 많은 경우의 overfitting 문제를 해결하는데 더 적합할 것이다. 이 결과에 따라 정규화 방법은 L2 로 선택하였다.


이후 과정에서는 모델의 성능과 overfitting 문제를 발생시킬 수 있는 activation function, layer 의 개수, batch size, epoch 값에 변화를 주면서 overfitting 문제가 발생하지 않으면서, 최고의 성능을 보이는 조건의 MLP 를 선정하였다.

### ·결론

조건 1개의 변화에 따라서도 overfitting 이 발생하기도 하고 overfitting 이 발생하지 않는 최고 성능의 모델이 나타나기도 하였다. 모델의 성능이 아무리 좋아도, overfitting 문제가 발생했을 가능성이 높다면, 새로운 데이터가 들어왔을 때, 분류를 제대로 하지 못한다. 특히 의료, 국방 분야와 같이 신속하고 정확한 판단을 내려하는 상황에서 overfitting 이 발생한 모델을 선택하는 것은 치명적일 수 있다. overfitting 문제의 중요성에 대해서 다시 한번 인식하게 되는 계기가 되었다.

실험을 하면서, 수업시간에 개념적으로 배웠던 모델들에 대해 직접 구현해보고, overfitting 에 영향을 줄 수 있는 요소들에 대해 구체적으로 알아볼 수 있었다. 또한, overfitting 되지 않는 최고 성능의 모델을 찾아냈다는 점에서 의의를 가진다.

실험을 하면서 아쉬웠던 부분은 MLP 의 여러 조건들 중 learning rate, hidden unit 처럼 성능과 overfitting 에 영향을 줄 수 있는 조건들이 더 많지만, 대표적인 4개만 선택했다는 점이다. 또한, 실습시간에 MLP 의 여러 layer 는 ReLU 를 사용하고, 마지막 output layer 를 sigmoid 를 썼을 때, 성능이 향상된 것을 관찰할 수 있었다. 이처럼 layer 내에서 여러 활성함수의 조합을 사용하면 성능이 어떻게 달라지는지도 관찰해보고 싶다.
